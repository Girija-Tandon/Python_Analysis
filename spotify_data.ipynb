```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/shreya_ghoshal_radio_tracks.csv')
df.sample(10)
```

```python
data = pd.read_csv("/content/top20_playlist_tracks.csv")
data.head(5)
```

```python
#Check for Missing Values
df.isnull().sum()
```

```python
#check basic info and summary
df.info()
```

```python
df.describe()
```

```python
# Top 5 Most Popular Tracks
data.sort_values(by='Popularity', ascending=False).head(5)[['Track Name', 'Artist(s)', 'Popularity']]
```

```python
#Plot Popularity Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=10, kde=True, color='skyblue')
plt.title('Popularity Distribution of Tracks')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#Convert Duration from ms to minutes
data['Duration (min)'] = df['Duration (ms)'] / 60000
```

```python
#Plot Duration vs Popularity
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Duration (min)', y='Popularity', hue='Release Year', palette='viridis')
plt.title('Track Duration vs Popularity')
plt.xlabel('Duration (minutes)')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

```python
#Number of Songs by Release Year
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Release Year', palette='Set2')
plt.title('Number of Tracks per Release Year')
plt.xticks(rotation=45)
plt.ylabel('Track Count')
plt.grid(True)
plt.show()
```

```python
#Total Playlist Duration in Minutes
total_duration_min = data['Duration (min)'].sum()
print(f"Total Playlist Duration: {total_duration_min:.2f} minutes")
```

```python
#Average Popularity
avg_popularity = df['Popularity'].mean()
print(f" Average Track Popularity: {avg_popularity:.2f}")
```

```python
# Group by Artist — Most Frequent Appearances
from collections import Counter

# Split multi-artist strings and count individual appearances
all_artists = data['Artist(s)'].str.split(', ').sum()
artist_counts = pd.Series(Counter(all_artists)).sort_values(ascending=False)

artist_counts.head(5)
```

```python
#Bar Plot – Top 5 Artists with Most Tracks
plt.figure(figsize=(8, 5))
artist_counts.head(5).plot(kind='bar', color='teal')
plt.title('Top 5 Most Frequent Artists in Playlist')
plt.xlabel('Artist')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#Popularity by Year (Trend)
yearly_avg = data.groupby('Release Year')['Popularity'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=yearly_avg, x='Release Year', y='Popularity', marker='o')
plt.title('Average Track Popularity by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Average Popularity')
plt.grid(True)
plt.show()
```

```python
#Correlation Heatmap (Numerical Fields)
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Duration (ms)', 'Popularity']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Identify Outlier Tracks in Popularity
q1 = df['Popularity'].quantile(0.25)
q3 = df['Popularity'].quantile(0.75)
iqr = q3 - q1

# Outliers: below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
outliers = data[(data['Popularity'] < (q1 - 1.5 * iqr)) | (data['Popularity'] > (q3 + 1.5 * iqr))]
outliers[['Track Name', 'Artist(s)', 'Popularity']]
```

```python
# Boxplot of Track Popularity
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Popularity'], color='lightblue')
plt.title("Boxplot of Track Popularity")
plt.xlabel("Popularity")
plt.grid(True)
plt.show()
```
